```python
data_without_runouts=data[data['dismissal_kind']!='run out']
data_without_runouts
```

```python
bowler_wickets=pd.DataFrame(data_without_runouts.groupby('bowler')['is_wicket'].sum()).reset_index().sort_values(by='is_wicket',ascending=False)
bowler_wickets
```

```python
data_without_extras=data[data['extra_runs']==0]
```

```python
data_without_extras
```

```python
no_of_ball=data_without_extras.groupby('bowler').size().reset_index(name='total_no_of_balls')
no_of_ball
```

```python
no_of_ball['overs']=np.floor(no_of_ball['total_no_of_balls']/6)
no_of_ball
```

```python
bowler_df=pd.merge(bowler_wickets,no_of_ball,on='bowler')
bowler_df
```

```python
bowler_df[bowler_df['overs']>10]
```

```python
no_of_runs_given=data.groupby('bowler')['total_runs'].sum().reset_index()
no_of_runs_given
```

```python
bowler_df=pd.merge(bowler_df,no_of_runs_given,on='bowler')
bowler_df
```

```python
bowler_df['economy']=bowler_df['total_runs']/bowler_df['overs']
bowler_df
```

```python
dot_balls_data=data[data['total_runs']==0]
dot_balls_data
```

```python
no_of_dot_balls=dot_balls_data.groupby('bowler').size().reset_index(name="dot_balls")
no_of_dot_balls
```

```python
bowler_df=pd.merge(bowler_df,no_of_dot_balls,on='bowler')
bowler_df
```

```python
bowler_df['%dot_balls']=(bowler_df['dot_balls']/bowler_df['total_no_of_balls'])*100
bowler_df
```

```python
bounderies=[4,6]
boundery_ball_data=data[data['batsman_runs'].isin(bounderies)]
boundery_ball_data
```

```python
boundery_balls=boundery_ball_data.groupby('bowler').size().reset_index(name="no_of_boundery_balls")
boundery_balls
```

```python
bowler_df
```

```python
bowler_df=pd.merge(bowler_df,boundery_balls,on='bowler')
bowler_df
```

```python
bowler_df['%boundery_balls']=(bowler_df['no_of_boundery_balls']/bowler_df['total_no_of_balls'])*100
bowler_df
```

```python
bowler_df['run_per_ball']=bowler_df['total_runs']/bowler_df['total_no_of_balls']
bowler_df
```

```python
bowler_no_of_matches=data.groupby('bowler')['id'].nunique().reset_index()
bowler_no_of_matches=bowler_no_of_matches.rename(columns={"id":"no_of_matches"})
bowler_no_of_matches
```

```python
bowler_df=pd.merge(bowler_df,bowler_no_of_matches,on='bowler')
bowler_df
```

```python
bowler_df['avg_wickets_per_match']=np.floor(bowler_df['is_wicket']/bowler_df['no_of_matches'])
bowler_df
```

```python
power_play_data=data[data['over']<=6]
power_play_data
```

```python
power_play_runs=power_play_data.groupby('bowler')['total_runs'].sum().reset_index(name="runs_given_in_powerplay")
power_play_runs
```

```python
power_play_bowler=power_play_data.groupby('bowler').size().reset_index(name="powerplay_balls")
power_play_bowler
```

```python
power_play_bowler['no_of_overs']=np.floor(power_play_bowler['powerplay_balls']/6)
power_play_bowler
```

```python
power_play_bowler=pd.merge(power_play_bowler,power_play_runs,on="bowler")
power_play_bowler
```

```python
power_play_bowler['powerplay_economy']=power_play_bowler['runs_given_in_powerplay']/power_play_bowler['no_of_overs']
power_play_bowler
```

```python
middle_over_data=data.query('over>6&over<17')
middle_over_data
```

```python
middle_over_data_without_extras=middle_over_data[middle_over_data['extra_runs']==0]
middle_over_balls=middle_over_data_without_extras.groupby('bowler').size().reset_index(name='no_of_balls_middle_over')
```

```python
middle_over_balls
```

```python
middle_over_runs=middle_over_data.groupby('bowler')['total_runs'].sum().reset_index()
middle_over_runs
```

```python
middle_over_bowler=pd.merge(middle_over_balls,middle_over_runs,on='bowler')
middle_over_bowler
```

```python
middle_over_bowler['middle_over_no_of_overs']=np.floor(middle_over_bowler['no_of_balls_middle_over']/6)
middle_over_bowler
```

```python
middle_over_bowler['middle_over_economy']=middle_over_bowler['total_runs']/middle_over_bowler['middle_over_no_of_overs']
middle_over_bowler
```

```python
death_over_data=data.query('over>16')
death_over_data
```

```python
death_over_data_without_extras=death_over_data[death_over_data['extra_runs']==0]
death_over_data_without_extras
```

```python
death_over_balls=death_over_data_without_extras.groupby('bowler').size().reset_index(name="no_of_balls_in_deathover")
death_over_balls
```

```python
death_over_runs=death_over_data.groupby('bowler')['total_runs'].sum().reset_index(name="runs_given_deathover")
death_over_runs
```

```python
death_over_bowler=pd.merge(death_over_balls,death_over_runs,on='bowler')
death_over_bowler
```

```python
death_over_bowler['no_of_over_deathover']=np.floor(death_over_bowler['no_of_balls_in_deathover']/6)
death_over_bowler
```

```python
death_over_bowler['deathover_ecnomy']=death_over_bowler['runs_given_deathover']/death_over_bowler['no_of_over_deathover']
death_over_bowler
```
